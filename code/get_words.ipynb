```python
# 状态列表与状态码相互转换
def status_transform2number(status=[0, 0, 0, 0, 0]):
    res = 0
    a = 1
    for i in status:
        res += a * i
        a = a * 3
    return res

def number_transform2status(number=0):
    res = []
    for i in range(5):
        res.append(number % 3)
        number //= 3
    return res

# 生成合法的状态排列
def generate_status():
    res = []
    for i in range(243):
        cur = []
        x = i
        for j in range(5):
            cur.append(x % 3)
            x //= 3
        res.append(cur)
        
    return res
```

```python
# 对猜测答案与正确答案之间进行验证
# 状态0代表完全错误，1代表字母正确且位置错误，2代表字母和位置都正确
def validation(guess, solution):
    result = [0, 0, 0, 0, 0]
    guess_list = list(guess)
    solution_list = list(solution)
    solution_letter_remainder = []
    
    for i in range(5):
        if guess_list[i] == solution_list[i]:
            result[i] = 2
        else:
            solution_letter_remainder.append(solution_list[i])
    
    for i in range(5):
        if result[i] == 2:
            continue
        for ch in solution_letter_remainder:
            if guess[i] == ch:
                result[i] = 1
                solution_letter_remainder.remove(ch)
                break
    
    return result

print( validation("grape", "hello") )
print( validation("heart", "hello") )
print( validation("hello", "hello") )
```

```python
# 在单词集中找出符合要求的所有单词
def get_legal_words(guess, status, valid_words):
    legal_words = []
    
    for word in valid_words:
        result = validation(guess, word)
        flag = True
        
        for i in range(5):
            if status[i] != result[i]:
                flag = False
                break
                
        if flag == True:
            legal_words.append(word)
    
    return legal_words


print( get_legal_words("hello", [2, 2, 0, 0, 0], valid_solutions) )
print( get_legal_words("hello", [2, 2, 0, 0, 1], valid_solutions) )
print( get_legal_words("hello", [2, 2, 0, 0, 1], valid_guesses) )
```

```python
# 计算信息量（信息熵）
def Entropy(p):
    if p == 0 or p == 1:
        return 0.0
    return - p * np.log(p) - (1.0 - p) * np.log(1.0 - p)


# 统计单词集中符合要求的所有单词的数量
def get_legal_words_cnt(guess, status, valid_words):
    legal_words = get_legal_words(guess, status, valid_words)
    return len(legal_words)


# 统计单词集中符合要求的所有单词的信息量（信息熵）
def get_legal_words_entropy(guess, status, valid_words):
    if len(valid_words) == 0:
        return 0.0
    p = get_legal_words_cnt(guess, status, valid_words) / len(valid_words)
    
    if p == 0 or p == 1:
        return 0.0
    entropy = Entropy(p)
    
    return entropy

# print( get_legal_words_entropy("hello", [2, 2, 0, 0, 0]) )
# print( get_legal_words_entropy("hello", [0, 0, 0, 0, 1]) )


# 计算总信息量（信息熵）
def get_legal_words_total_entropy(guess, last_status, valid_words):
    total_status = generate_status()
    total_entropy = 0
    for status in total_status:
        total_entropy += get_legal_words_entropy(guess, status, valid_words)
    return total_entropy

# print( get_legal_words_total_entropy("hello") )


# 获取词汇频率
def get_word_frequency(guess):
    result = word_frequency[ word_frequency["word"] == guess ]
    weight = result["frequency"].tolist()
    res = sum(weight)
    
    # print(res)
    return res
    
print( get_word_frequency("hello") )
```

```python
guess_legal_words_data = []
guess_legal_words_metrics_data = []

for guess in valid_guesses[0:0]:
    total_entropy = 0
    word_frequency = get_word_frequency(guess)
    # print(guess)
    
    for number in range(243):
        legal_words = get_legal_words(guess, number_transform2status(number), valid_solutions)
        len_of_legal_words = len(legal_words)
        if len_of_legal_words == 0:
            continue
        guess_legal_words_data.append([guess, number, legal_words, len_of_legal_words])
        total_entropy += Entropy(len_of_legal_words / valid_solutions_cnt)
    
    guess_legal_words_metrics_data.append([guess, total_entropy, word_frequency])
```

```python
guess_legal_words_data = pd.DataFrame(guess_legal_words_data, columns=["guess", "number", "legal_words", "len_of_legal_words"], dtype=int)
guess_legal_words_data.to_csv("guess_legal_words_data(valid_solutions).csv")
guess_legal_words_metrics_data = pd.DataFrame(guess_legal_words_metrics_data, columns=["guess", "total_entropy", "word_frequency"], dtype=float)
guess_legal_words_metrics_data.to_csv("guess_legal_words_metrics_data(valid_solutions).csv")
```
