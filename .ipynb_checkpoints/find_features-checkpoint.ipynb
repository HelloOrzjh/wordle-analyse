```python
# 计算两个单词之间的相似度(欧拉距离)
def euler_distance(vec_x, vec_y):
    dist = 0
    for i in range(26):
        dist += (vec_x[i] - vec_y[i]) * (vec_x[i] - vec_y[i])
    dist = np.sqrt(dist)
    return dist
```

```python
# one-hot对单词进行编码
def one_hot(x):
    vec_x = np.zeros(26)
    
    for i in x:
        vec_x[ ord(i) - ord('a') ] += 1
        
    return vec_x

one_hot("hello")
```

```python
Problem_C_Data_Wordle["one_hot"] = Problem_C_Data_Wordle["word"].apply(one_hot)
Problem_C_Data_Wordle
```

```python
col = ["word"]
for i in range(26):
    col.append( chr(97 + i) ) 
one_hot_df = pd.DataFrame([], columns=col)

word = Problem_C_Data_Wordle["word"].tolist()
data = Problem_C_Data_Wordle["one_hot"].tolist()

for i in range( len(data) ):
    data[i] = list(data[i])
    cur = [ word[i] ] + data[i]
    # print(cur)
    one_hot_df.loc[ len(one_hot_df.index) ] = cur

one_hot_df
```

```python
import scipy.cluster.hierarchy as sch  #层次聚类
 
A = one_hot_df.iloc[:, 1:]
Z = sch.linkage(A, method = 'single', metric = 'euclidean') #euclidean代表欧式距离，#single代表简单连接
 
#将层次聚类结果以树状图表示出来
fig=plt.figure(figsize=(14,10))
sch.dendrogram(Z, 0)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering #导入sklearn的层次聚类函数

data1 = A

clf1 = AgglomerativeClustering(n_clusters = 2, linkage = 'ward')
s = clf1.fit(data1)
pred1 = clf1.fit_predict(data1)
score1 = silhouette_score(data1, pred1)
print(f'聚类轮廓系数为：{score1}')
pca = PCA(n_components=2)  # 输出两维
newData1 = pca.fit_transform(data1)  # 载入N维

x, y = [], []
for i in range(7):
    x.append([])
    y.append([])
    
for index, value in enumerate(pred1):
    x[value].append(newData1[index][0])
    y[value].append(newData1[index][1])
plt.figure(figsize=(10, 10))

# #定义坐标轴
k = 200
for i in range(7):
    plt.scatter(x[i], y[i], s=k)
plt.legend(['d1','d2','d3','d4','d5','d6','d7'])
plt.show()
```
