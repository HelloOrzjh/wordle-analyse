```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
data = pd.read_excel("Problem_C_Data_Wordle.xlsx", header=1)
data = data.drop(["Unnamed: 0"], axis=1)
data = data.sort_values(by="Date", axis=0, ascending=True)
data
```

```python
data.sort_values(by="Number_of_reported_results", axis=0, ascending=False)
```

```python
data.sort_values(by="Number_in_hard_mode", axis=0, ascending=False)
```

```python
data.sort_values(by="1 try", axis=0, ascending=False)
```

```python
data["Total_percent"] = data.iloc[:, 5:12].sum(axis=1)

data["Average_Tries"] = data.iloc[:, 5] * 1 + data.iloc[:, 6] * 2 + \
                        data.iloc[:, 7] * 3 + data.iloc[:, 8] * 4 + \
                        data.iloc[:, 9] * 5 + data.iloc[:, 10] * 6 + data.iloc[:, 11] * 7 
data["Average_Tries"] = data["Average_Tries"] / data.iloc[:, 12]
data.sort_values(by="Average_Tries", axis=0, ascending=True)
```

```python
data["Average_Tries"].mean()
```

```python
x = data["Average_Tries"].tolist()

plt.figure(figsize=(12,6),dpi=100)
plt.hist(x,bins=20)
plt.show()
```

```python
alphabet = []
for i in range(97, 123):
    alphabet.append([chr(i), 0, 0, 0])
alphabet = pd.DataFrame(alphabet, columns=["Letter", "Appear_Count", "Total_Average_Tries", "Average_Tries"], dtype=int)
, 
for i in range(data.shape[0]):
    row = data.loc[i, :]
    Word = row["Word"].replace(' ', '')
    Word_set = set(Word)
    Number_of_reported_results = row["Number_of_reported_results"]
    Number_in_hard_mode = row["Number_in_hard_mode"]
    Average_Tries = row["Average_Tries"]
    
    for letter in Word_set:
        if ord(letter) - 97 >= 26 or ord(letter) - 97 < 0:
            print(Word)
        alphabet.iloc[ord(letter) - 97, 1] += 1
        alphabet.iloc[ord(letter) - 97, 2] += Average_Tries
    
alphabet["Average_Tries"] = alphabet["Total_Average_Tries"] / alphabet["Appear_Count"]

alphabet.iloc[:,1].sum()
alphabet.sort_values(by="Appear_Count", axis=0, ascending=True)
```

```python
sorted_alphabet = alphabet.sort_values(by="Appear_Count", axis=0, ascending=True)

plt.figure(figsize=(12, 5))
plt.barh(sorted_alphabet.iloc[:,0], sorted_alphabet.iloc[:,1])
plt.show()
```

```python
alphabet_distribution = []
for i in range(97, 123):
    alphabet_distribution.append([chr(i), 0, 0, 0, 0, 0])
alphabet_distribution = pd.DataFrame(alphabet_distribution, columns=["Letter", "pos1", "pos2", "pos3", "pos4", "pos5"], dtype=int)
 
for i in range(data.shape[0]):
    row = data.loc[i, :]
    Word = row["Word"].replace(' ', '')
    Word_set = set(Word)
    Number_of_reported_results = row["Number_of_reported_results"]
    Number_in_hard_mode = row["Number_in_hard_mode"]
    Average_Tries = row["Average_Tries"]
    
    if len(Word) != 5:
        continue
    
    pos = 1
    for letter in Word:
        if ord(letter) - 97 >= 26 or ord(letter) - 97 < 0:
            print(Word)
            break
        if pos > 5:
            print(Word)
            break
        
        alphabet_distribution.iloc[ord(letter) - 97, pos] += 1
        pos += 1
    
alphabet_distribution
```
